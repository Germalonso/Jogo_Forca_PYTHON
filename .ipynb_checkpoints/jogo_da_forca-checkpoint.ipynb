```python
import random 
from os import system, name

def limpa_tela():
    if name == 'nt':
        _ = system('cls')
    else:
        _ = system('clear')

def escolher_palavra():
    palavras = ['in alfredo pereira']
    return random.choice(palavras)

def print_github_germalonso():
    germalonso = r"""
    _____                               _
   / ____|                             | |
  | |  __  ___  _ __  _ __ ___    __ _ | |  ___   _ __    ____    ___  
  | | |_ |/ _ \| '__|| '_ ` _ \  / _` || | / _ \ | '_ \  //  \\  / _ \ 
  | |__| |  __/| |   | | | | | || (_| || || (_) || | | | \\___  | (_) |
   \_____/\___||_|   |_| |_| |_| \__,_||_| \___/ |_| |_| \\__//  \___/
                                                         
    """
    
    
    print(germalonso)

def mostrar_palavra(palavra, letras):
    return ' '.join([letra if letra in letras or letra == ' ' else '_' for letra in palavra])

def mostrar_forca(erros):
    estagios = [
        """
           -----
           |   |
               |
               |
               |
               |
         =========
        """,
        """
           -----
           |   |
           O   |
               |
               |
               |
         =========
        """,
        """
           -----
           |   |
           O   |
           |   |
               |
               |
         =========
        """,
        """
           -----
           |   |
           O   |
          /|   |
               |
               |
         =========
        """,
        """
           -----
           |   |
           O   |
          /|\\  |
               |
               |
         =========
        """,
        """
           -----
           |   |
           O   |
          /|\\  |
          /    |
               |
         =========
        """,
        """
           -----
           |   |
           O   |
          /|\\  |
          / \\  |
               |
         =========
        """
    ]
    print(estagios[erros])
    
def jogo_da_forca():
    limpa_tela()
    palavra = escolher_palavra()
    tentativas = 6
    letra_erradas = []
    letras = []
    print_github_germalonso()
    print("===================== Bem vindo ao jogo da forca =====================")
    print("")
    while(tentativas > 0):
        print(f"\nPalavra: ",mostrar_palavra(palavra,letras))
        mostrar_forca(6-tentativas)
        while True:
            tentativa = input("Digite uma letra: ").lower()
            if len(tentativa) == 1 and tentativa.isalpha():
                break  # Sai do loop se for uma letra válida
            else:
                print("Entrada inválida! Digite apenas uma letra.\n")
        
        if tentativa in letras or tentativa in letra_erradas:
            print("Você já tentou essa letra.\n")
            continue
            
        if tentativa in palavra:
            letras.append(tentativa)
            if all(letra in letras for letra in palavra):
                print(f"Parabéns !!!!!!!!! você acertou a palavra: {palavra}")
                escolha = input("\nDeseja jogar novamente? (s/n): ").lower()
                if escolha == 's':
                    jogo_da_forca()
                else:
                    break
        else:
            letra_erradas.append(tentativa)
            tentativas -= 1
            
        if tentativas == 0:
            print(f"Você perdeu a palavra era {palavra}")
            escolha = input("\nDeseja jogar novamente? (s/n): ").lower()
            if escolha == 's':
                jogo_da_forca()
            else:
                break
            

jogo_da_forca()
    
```
